## Forecasting fertility rate of India 

```python
#Let us check if the model works properly
df_india=pd.DataFrame(data=df.index,columns=['India'])
df_india['India']=df['India']
df_india.index.name='Year'
df_india.columns=["India"]
df_india
```

We will use time series model for forecasting.

```python
df_india_2000=df_india.loc[:2000,:]

import warnings
warnings.filterwarnings('ignore')

import statsmodels.api as smapi

model = smapi.tsa.arima.ARIMA(df_india_2000, order=(1,1,2))

result = model.fit()



#making predictions
predictions = model.fit().forecast(steps=20)

#predictions
n=2001
for prediction in predictions:
    df_india_2000.append(pd.Series(name=n))
    df_india_2000.at[n,'India']=prediction
    n=n+1
```

```python

ax = df_india_2000.loc[2000:2020,'India'].plot(x='Year', y='India',color='g',marker = "o", markerfacecolor = 'g')
df_india.loc[2000:2020,'India'].plot(ax=ax, x='Year', y='India',linestyle='dotted',marker = "o", markerfacecolor = 'b',figsize=(17,9))
plt.xlabel("Year")
plt.ylabel("Fertility rate")
ax.set_title("Prediction using ARIMA");
print("\033[0;32mPredicted\033[00m")
print("\033[0;34mOriginal\033[00m")
```

```python
from sklearn.metrics import mean_squared_error
  
MSE_arima = mean_squared_error(df_india.loc[2000:2020,'India'],df_india_2000.loc[2000:2020,'India'])
MSE_arima
```

### Forecasting death rate of India till 2050 

Time series models rely on historical patterns and trends in the data to make predictions, but if there is a sudden, significant change in the data, such as the spike in deaths caused by the COVID-19 pandemic, it may not be accurately captured by the model.
In such cases, polynomial regression can be a useful alternative for forecasting

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

model = LinearRegression()
poly_reg = PolynomialFeatures(degree=2)

dth_india_2020 = dth_india
for n in range(2021, 2051):
    x = dth_india_2020.index.values.reshape(-1, 1)
    x_poly = poly_reg.fit_transform(x)
    y = dth_india_2020['India']
    model.fit(x_poly, y)
    z = model.predict(poly_reg.fit_transform([[n]]))
    new_row = pd.DataFrame({'India': [z]}, index=[n])
    dth_india_2020 = pd.concat([dth_india_2020, new_row])
```

```python
dth_india_2020
```

```python
dth_india_2020 = dth_india_2020.astype(float)
ax = dth_india_2020.loc[1960:2020,'India'].plot(x='Year', y='India',color='g')
dth_india_2020.loc[2020:2050,'India'].plot(ax=ax, x='Year', y='India',linestyle='dotted',color='r',figsize=(17,9))
plt.xlabel("Year")
plt.ylabel("Death Rate")
ax.set_title("Death Rate Prediction till 2050");
print("\033[0;32mOriginal data\033[00m")
print("\033[0;31mPredictions\033[00m")
```

## By polynomial regression

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

model = LinearRegression()
poly_reg = PolynomialFeatures(degree=4)

mg_india_2020 = mg_india.loc[:2020,:]
for n in range(2021, 2051):
    x = mg_india_2020.index.values.reshape(-1, 1)
    x_poly = poly_reg.fit_transform(x)
    y = mg_india_2020['India']
    model.fit(x_poly, y)
    z = model.predict(poly_reg.fit_transform([[n]]))
    new_row = pd.DataFrame({'India': [z]}, index=[n])
    mg_india_2020 = pd.concat([mg_india_2020, new_row])
```

```python
def det(n):
    return dth_india_2020.at[n, 'India'] 
def mig(n):
    return mg_india_2020.at[n, 'India']
```

```python
#define c in terms of x 
#c=35*0.222*x/1000
```

Finding population in 2030, with constant fertility rate 2.0 from current time to 2030

```python
from sympy import symbols, solve, I

#n = symbols('n', integer=True, nonnegative=True)
#c = symbols(' c', real=True, nonnegative=True)
c = ((2*0.222*35)/1000)
p=1370004385;
y=1370004385;
n=2030
# Update the value of y based on the data from the DataFrame dth_india
for i in range(2021, n+1):
    y = y * (1+c - (det(i-1)/1000));


for i in range(2021,n):
    a=1;
    for j in range(i+1, n):
        a = a * (1+c - (det(j-1)/1000));
    y = y + mig(i)* a;

```
